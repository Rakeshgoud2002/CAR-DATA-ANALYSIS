```python
df = np.array([35, 25, 25, 15])
mylabels = ["name","engine","max_power","seats"]

myexplode= [0.2, 0, 0, 0]
mycolors = ["black", "hotpink", "b", "#4CAF50"]
plt.pie(df, labels = mylabels, explode = myexplode,shadow = True, colors=mycolors)
plt.legend()
plt.show() 
```
